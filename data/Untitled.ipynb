```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv('네이버맛집_전처리_3.csv', header='infer', engine='python', encoding='cp949', sep=',')
```

```python
for i, url in enumerate(df.loc[:,'images_url']):
    urls = []
    for sp in url.split(','):
#         print(re.sub("]|\[|'", "",sp).strip())
        urls.append([re.sub("]|\[|'", "", sp).strip()])
    df.loc[i, 'images_url_preprocess'] = urls
```
